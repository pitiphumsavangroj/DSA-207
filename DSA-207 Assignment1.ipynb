```python
url = 'https://raw.githubusercontent.com/ferdinjoe/Jupyter_Data_Visualization/master/kanjanaphisek_bangkok_pm25.csv'
df = pd.read_csv(url)
df
```

```python
df.rename(columns={'UTC Hour':'Hour'}, inplace=True)
df.head()
```

```python
df['datetime'] = pd.to_datetime(df[['Year', 'Month', 'Day', 'Hour']]) + pd.Timedelta(hours=7) # UTC to Bangkok time
df.head()
```

```python
df.index = df['datetime']
df.head()
```

```python
df.Hour = df.index.hour
df.head()
```

```python
df.drop(columns=['PM10_mask','Retrospective','datetime'],inplace=True)
df.head()
```

```python
df['PM2.5'].plot(figsize=(14,6),title="Kanjanaphisek Rd, Bangkok PM2.5 (ug/m3) 2016/03 - 2019/02/7", color='blue', grid=True)
df.shape
```

```python
pv = pd.pivot_table(df, index=df.index.dayofyear, columns=df.index.year,values='PM2.5', aggfunc='mean')
pv
```

```python
pv.plot(title="Bangkok PM2.5 Daily Average Yearly comparison (ug/m3)",figsize=(14,6), grid=True)
```

```python
pv2 = pd.pivot_table(df, index=df.index.dayofyear, columns=df.index.year, values='PM2.5', aggfunc='max')
pv2
```

```python
pv2.plot(title="Bangkok PM2.5 Daily Maximum Yearly comparison (ug/m3)",figsize=(14,6), grid=True)
```

```python
year = 2016

fig, ax = plt.subplots(4, figsize=(12, 20))
for i in range(0, 4):
    x = df[df['Year'] == year+i].index
    y = df[df['Year'] == year+i]['PM2.5']

    ax[i].scatter(x, y, c=y, s=y*2, alpha=0.5, linewidths=1, marker='.')
    ax[i].grid(True)
    ax[i].set_title("PM2.5 in Year "+str(year+i))

plt.show()
```

```python
year = 2016

fig, ax = plt.subplots(4, figsize=(12, 20))
for i in range(0, 4):
    x = pv2[year+i].index
    y = pv2[year+i]

    ax[i].scatter(x, y, c=y, s=y*10, alpha=0.5, linewidths=1, marker='.')
    ax[i].grid(True)
    ax[i].set_title("PM2.5 in Year "+str(year+i))
    
    x_pos = [*pv2[pv2[year+i] == y.max()].index][0]
    y_pos = y.max()
    ax[i].annotate("("+str(x_pos)+", "+str(y_pos)+")", (x_pos, y_pos+0.5))

plt.show()
```

```python
level = []
for row in df['PM2.5']:
    if row <= 50:
        level.append('normal')
    elif row <= 100:
        level.append('high')
    else:
        level.append('very high')

df['level'] = level
df.head()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df[df['level']=='very high'].index, df[df['level']=='very high']['PM2.5'], 
            s=df[df['level']=='very high']['PM2.5'], marker='.')
plt.scatter(df[df['level']=='high'].index, df[df['level']=='high']['PM2.5'], 
            s=df[df['level']=='high']['PM2.5'], marker='.')
plt.scatter(df[df['level']=='normal'].index, df[df['level']=='normal']['PM2.5'], 
            s=df[df['level']=='normal']['PM2.5'], marker='.')
plt.grid(True)

plt.show()
```

```python
fraction = []
fraction.append(df[df['level']=='very high']['level'].count())
fraction.append(df[df['level']=='high']['level'].count())
fraction.append(df[df['level']=='normal']['level'].count())
print(fraction)

labels = ["very high", "high", "normal"]
explode = [0.2, 0.2, 0.2]

plt.pie(fraction, labels=labels, explode=explode, autopct="%1.1f%%")
plt.legend(title="PM2.5", bbox_to_anchor=(1, -0.3, 0.7, 1))

plt.show()
```
