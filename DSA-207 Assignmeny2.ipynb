```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='winter')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/customers.csv")
print(data.head())
```

```python
data["Income"] = data[["Annual Income (k$)"]]
data["Spending"] = data[["Spending Score (1-100)"]]
data = data[["Income", "Spending"]]
print(data.head())
```

```python
from sklearn.cluster import MeanShift

model = MeanShift(bandwidth=10)
model.fit(data)
pred = model.fit_predict(data)
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
plt.scatter(data["Income"], data["Spending"], c=pred, cmap='rainbow', alpha=0.9)
plt.show()
```

```python
df = pd.read_csv('fishdata.csv', encoding='unicode_escape', header=None)
df.head()
```

```python
mylist = [*range(12,35)]
fish = df.drop(columns=mylist)

header = ['year', 'month', 'day', 'hour', 'minute', 'second', 'carbon monoxide', 'PM 2.5', 'external temp', 'humidity', 'internal temperature', 'ignore']
fish.columns = header
fish = fish.dropna()
fish.head()
```

```python
fish['year'] = fish['year'].astype('int64')
fish['month'] = fish['month'].astype('int64')
fish['minute'] = fish['minute'].astype('int64')
fish['second'] = fish['second'].astype('int64')
fish['carbon monoxide'] = pd.to_numeric(fish['carbon monoxide'], errors='coerce')
fish['carbon monoxide'] = fish['carbon monoxide'].astype('float64')
fish.dtypes
```

```python
fish.describe()
```

```python
fish = fish[fish['year'] > 2000]
fish = fish[fish['year'] <= 2021]
fish.describe()
```

```python
fish = fish[fish['external temp'] > 0]
fish.describe()
```

```python
fdata = fish.head(30000)
```

```python
from sklearn.cluster import MeanShift

model = MeanShift(bandwidth=50)
model.fit(fdata)
pred = model.fit_predict(fdata)

plt.figure(figsize=(12, 10))
plt.scatter(fdata['external temp'], fdata['humidity'], c=pred, cmap='plasma', alpha=0.9)
plt.show()
```

```python
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection='3d')

xdata = fish['internal temperature']
ydata = fish['external temp']
zdata = fish['PM 2.5']

ax.scatter(xdata, ydata, zdata, c=zdata, cmap='autumn')
ax.set_xlabel('internal temperature')
ax.set_ylabel('external temperature')
ax.set_zlabel('PM 2.5')
ax.set_title('Scatter3D')
```

```python
Z = fish.pivot_table(index='internal temperature', columns='external temp', values='PM 2.5').T.values
X_unique = np.sort(fish['internal temperature'].unique())
Y_unique = np.sort(fish['external temp'].unique())
X, Y = np.meshgrid(X_unique, Y_unique)

fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection='3d')
ax.contour(X, Y, Z, 50, cmap='winter')
ax.set_xlabel('internal temperature')
ax.set_ylabel('external temperature')
ax.set_zlabel('PM 2.5')
ax.set_title('Contour3D')
```
